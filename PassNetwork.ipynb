```python
import pandas as pd
from mplsoccer.pitch import Pitch
```

```python
#read in xlsx
df = pd.read_excel('England_Premier-League_2022-2023_Liverpool_vs_Arsenal.xlsx')

df
```

```python
#filter df to get only the team we want
df = df[df['team']=='Liverpool']
```

```python
#now we want to find our passes and recipients and then filter for only passes
df['passer'] = df['playerName']
df['recipient'] = df['playerName'].shift(-1)

#find passes and then only look for the successful passes
passes = df[df['type']=='Pass']
successful = passes[passes['outcomeType']=='Successful']
```

```python
successful
```

```python
#find the first subsititution and filter the successful dataframe to be less than that minute
subs = df[df['type']=='SubstitutionOff']
subs = subs['minute']
firstSub = subs.min()

successful = successful[successful['minute'] < firstSub]
```

```python
successful
```

```python
#now we need to find the average locations and counts of the passes
average_locations = successful.groupby('passer').agg({'x':['mean'],'y':['mean','count']})
average_locations.columns = ['x','y','count']
```

```python
#now we need to find the average locations and counts of the passes
average_locations_recipient = successful.groupby('recipient').agg({'x':['mean'],'y':['mean','count']})
average_locations_recipient.columns = ['x','y','count']
```

```python
average_locations_recipient
```

```python
average_locations
```

```python
successful
```

```python
#now we need to find the number of passes between each player
pass_between = successful.groupby(['passer','recipient']).playerName.count().reset_index()
pass_between.rename({'playerName':'pass_count'},axis='columns',inplace=True)

#merge the average location dataframe. We need to merge on the passer first then the recipient
pass_between = pass_between.merge(average_locations, left_on='passer',right_index=True)
pass_between = pass_between.merge(average_locations, left_on='recipient',right_index=True,suffixes=['', '_end'])
```

```python
pass_between
```

```python
#set minimum threshold of combinations.. I like 5 for high passing teams. 2 or 3 for low passing.
pass_between = pass_between[pass_between['pass_count']>2]
```

```python
pass_between
```

```python
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch

# set up the figure and subplot
fig, ax = plt.subplots(figsize=(13, 8))

# create the pitch
pitch = Pitch(pitch_type='statsbomb', pitch_color='#000000', line_color='#c7d5cc',
              pitch_length=8, pitch_width=13)

# plot the pitch
pitch.draw(ax=ax)
ax.invert_yaxis()  # invert the y-axis of the plot

# plot the arrows and nodes
arrows = pitch.arrows(1.2*pass_between.x,.8*pass_between.y,1.2*pass_between.x_end,.8*pass_between.y_end,
                     width = 3, headwidth = 3, color = '#12F0C8', ax=ax, zorder = 1, alpha = .5)
nodes = pitch.scatter(1.2*average_locations.x,.8*average_locations.y,
                     s = 300, color = '#12F0C8', edgecolors = 'black', linewidth = 2.5, alpha = 1, zorder = 1, ax=ax)

average_locations['playerName'] = average_locations.index
for i, row in average_locations.iterrows():
    ax.annotate(row.name, xy=(1.2*row['x'], .8*row['y'] + 2.5), ha='center', va='center', color='white', fontsize=12)
```
